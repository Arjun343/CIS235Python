```python
# Question 1:
# 1. Create a new DataFrame out of the following actual crime data from 2015:
#    NOTE: There are many ways you can do this. for example:
#          First add the data and headers to a dictionary and then build a DataFrame from the dictionary
#          Create the DataFrame with column names, then set the index, then add the data row-by-row.
#          Several others
#    Choose any method you want that works. However, some valid techniques are much harder/easier than others.
#    Try to think of the easiest way to accomplish this task

# agency_code	location	population	homicides	assaults	robberies
# UT01803	Salt Lake City, UT	191992	8	874	469
# TXSPD00	San Antonio, TX	1463586	94	5465	1986
# CA03711	San Diego, CA	1400467	37	3601	1378
# MOSPD00	St. Louis, MO	317095	188	3521	1790
# NY05101	Suffolk County, NY	1341453	24	895	677

# 2. Set the agency_code as the index
# 3. Output the first 2 records in the "nice looking" Pandas DataFrame format (i.e. not the default print() format)

import pandas as pd

# Create a list of dictionaries with the data
data = [
    {'agency_code': 'UT01803', 'location': 'Salt Lake City, UT', 'population': 191992, 'homicides': 8, 'assaults': 874, 'robberies': 469},
    {'agency_code': 'TXSPD00', 'location': 'San Antonio, TX', 'population': 1463586, 'homicides': 94, 'assaults': 5465, 'robberies': 1986},
    {'agency_code': 'CA03711', 'location': 'San Diego, CA', 'population': 1400467, 'homicides': 37, 'assaults': 3601, 'robberies': 1378},
    {'agency_code': 'MOSPD00', 'location': 'St. Louis, MO', 'population': 317095, 'homicides': 188, 'assaults': 3521, 'robberies': 1790},
    {'agency_code': 'NY05101', 'location': 'Suffolk County, NY', 'population': 1341453, 'homicides': 24, 'assaults': 895, 'robberies': 677}
]

# Create the DataFrame directly from the list of dictionaries and set 'agency_code' as the index
crime_df = pd.DataFrame(data).set_index('agency_code')

# Output the first 2 records in a nice-looking format
print(crime_df.head(2))
```

```python
# Question 2

# 1. Print just the location column
# 2. Print the entire row representing Salt Lake City's data using the .loc[] property
# 3. Print each of the column names/headers one-at-a-time


print(crime_df['location'])
print('\n')
print(crime_df.loc['UT01803'])
print('\n')
for column in crime_df.columns:
    print(column)
```

```python
# Question 3

# 1. Add a new record for:
#       FL02902, Tampa, FL, 364383, 27, 2730, 733
# 2. Create a new DataFrame that includes the homicides per capita for each location
#       This is calculated as homicides / population * 100000 (round to zero decimal places)
#       You can use a calculator to generate these numbers and hard-code them into a DataFrame
# 3. Combine that new DataFrame with the prior existing DataFrame
#       This could be done in multiple ways. Take your pick
# 4. Print the entire combined DataFrame in Pandas format (i.e. not using print())
#
# NOTE: an easier way to accomplish the same task would be to create a new, calculated column in the
# existing DataFrame. But that is not what is being tested here. So perform the steps as described above.

import pandas as pd
# 1. Add a new record for Tampa, FL
new_record = {'agency_code': 'FL02902', 'location': 'Tampa, FL', 'population': 364383, 'homicides': 27, 'assaults': 2730, 'robberies': 733}
crime_df = crime_df.append(new_record, ignore_index=True)

# 2. Create a new DataFrame for homicides per capita
homicides_per_capita = (crime_df['homicides'] / crime_df['population'] * 100000).round(0)

# 3. Combine the new DataFrame with the existing one
crime_df['homicides_per_capita'] = homicides_per_capita

# 4. Print the entire combined DataFrame in Pandas format
crime_df
```

```python
# Question 4

# 1. Sort the DataFrame by homicides per capita from greatest to least on the original DataFrame (not a copy)
# 2. Filter/reduce the DataFrame down to only those with more than 50 total homicides OR (not 'and') over 5000 assaults
# 3. Print the DataFrame

# 1. Sort the DataFrame by homicides per capita from greatest to least
crime_df = crime_df.sort_values(by='homicides_per_capita', ascending=False)

# 2. Filter the DataFrame based on the given conditions
filtered_df = crime_df[(crime_df['homicides'] > 50) | (crime_df['assaults'] > 5000)]

# 3. Print the resulting DataFrame
filtered_df
```
